1. Difference between "Love" and "love" in NLP

   
NLP models treat words as tokens.
Without normalization:
"Love" and "love" are treated as different tokens
This increases vocabulary size unnecessarily.
Converting to lowercase ensures:
Better generalization
Reduced sparsity

2. What happens if stopwords are not removed?

   
Adds noise to data
Increases dimensionality
Slows down model training
Reduces model performance in tasks like classification

3. When removing stopwords is harmful

Scenario 1: Sentiment Analysis

Sentence: "I am not happy"
Removing "not" → "I am happy"  (wrong meaning)

Scenario 2: Question Answering

Sentence: "What is the capital of India?"
Removing stopwords → loses structure and meaning

Task 2: Preprocessing Function

```python
import re

def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return []

    # Remove URLs and emails
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Convert to lowercase
    text = text.lower()

    # Handle repeated characters (soooo → soo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters / emojis
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    # Remove very short tokens (<=2), but keep 'no', 'not'
    filtered_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            filtered_tokens.append(word)

    return filtered_tokens
```

Task 3: Stress Testing

```python
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for text in sample_inputs:
    tokens = preprocess_text(text)
    cleaned_sentence = " ".join(tokens)

    results.append({
        "original": text,
        "tokens": tokens,
        "cleaned": cleaned_sentence
    })

for r in results:
    print("Original:", r["original"])
    print("Tokens:", r["tokens"])
    print("Cleaned:", r["cleaned"])
    print("-" * 50)
```

Task 4: Token Analytics

```python
import numpy as np

for r in results:
    tokens = r["tokens"]

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(t) for t in tokens]) if tokens else 0

    print("Sentence:", r["original"])
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Avg Token Length:", round(avg_length, 2))
    print("-" * 50)
```

Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []
for r in results:
    all_tokens.extend(r["tokens"])

counter = Counter(all_tokens)

print("Top 10 frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[-5:])
```

Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        sentence = " ".join(tokens)

        all_tokens.extend(tokens)
        clean_sentences.append(sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
output = full_pipeline(sample_inputs)
print(output)
```
